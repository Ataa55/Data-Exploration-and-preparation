```python
from selenium import webdriver
import pandas as pd
```

```python
mohsen = webdriver.Chrome()
```

```python
mohsen.get("https://www.amazon.com/s?k=mopile+phones+samsung&crid=1Y3A3CKWDCZ62&sprefix=mobile+phones+samsun%2Caps%2C325&ref=nb_sb_noss_2")
```

```python
objects = mohsen.find_elements("xpath",'//div[@class="s-card-container s-overflow-hidden aok-relative puis-include-content-margin puis puis-v6rsvs73r5ieo2jbx36zk3g8bu s-latency-cf-section s-card-border"]')
```

```python
for i in objects:
    print(i.text)
```

```python
names = []
prices = []
rates = []
stars = []
images = []
for element in objects:
    #############
    try:        
        names.append(element.find_element("xpath", './/span[@class="a-size-medium a-color-base a-text-normal"]').text)
    except:
        names.append("null")
    ####################3
    try:
        prices.append(element.find_element("xpath", './/span[@class="a-price-whole"]').text)
    except:
        prices.append("null")
   #################
    try:
        rates.append(element.find_element("xpath", './/span[@class="a-size-base s-underline-text"]').text)
    except:
        rates.append("null")
```

```python
print(len(names))
print(len(prices))
print(len(rates))
```
